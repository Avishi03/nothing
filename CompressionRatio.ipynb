```python
import matplotlib.pyplot as plt
import random

# Example data (simulating a scenario with more chunks for better representation)
chunk_indices = list(range(100))  # Simulating 100 chunks
uncompressed_sizes = [random.randint(800, 1300) for _ in range(100)]  # Simulated uncompressed sizes between 800 and 1300 bytes
compressed_sizes = [random.randint(500, 1000) for _ in range(100)]  # Simulated compressed sizes between 500 and 1000 bytes

# Calculate the compression ratio for each chunk (uncompressed_size / compressed_size)
compression_ratios = [u / c if c > 0 else 0 for u, c in zip(uncompressed_sizes, compressed_sizes)]

# Plot the compression ratio graph
plt.figure(figsize=(10, 6))
plt.plot(chunk_indices, compression_ratios, marker='o', linestyle='-', color='b', label='Compression Ratio')
plt.xlabel('Chunk Index')
plt.ylabel('Compression Ratio')
plt.title('Compression Ratio for Each Chunk')
plt.xticks(range(0, 101, 10))  # Show x-axis labels every 10 chunks (for clarity)
plt.grid(True)
plt.legend()
plt.show()
```
